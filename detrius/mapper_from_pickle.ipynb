```python
import pandas as pd
import folium
import fiona
import geopandas as gpd
import requests
import numpy as np
from bs4 import BeautifulSoup
from pandas.io.json import json_normalize
from shapely.geometry import Polygon
from math import pi, sin, cos
from geojson import GeoJSON
import json
import matplotlib

%matplotlib inline

points_frame=pd.read_pickle('points_pickle')
points_frame.head()
# hex_frame.plot(column='score')
# hex_frame.crs = {'init': 'epsg:3857'}


# hex_json = json.loads(hex_frame[['geometry','index1']].to_json())
# with open('hexes.json', 'w') as outfile:
#     json.dump(hex_json, outfile)
```

```python
points_frame['score'] = points_frame['scores'].apply(lambda x: x['score'])

points_frame.head()
```

```python
trim = points_frame[['score', 'shape_id']].groupby('shape_id').mean().reset_index()
trim.head()
```

```python
cities = ['Washington']

"""cities = ['Washington',
'San_Francisco',
'New_York']"""



wider_map = folium.Map()



for city in cities:
    file_name = city+"_hex.json"
    json_data=open(file_name).read()
    hexes = json.loads(json_data)
    
    wider_map.choropleth(geo_data=hexes,
                data=trim,
                columns=['shape_id', 'score'],
                key_on='feature.id',
                fill_color='YlGn',
                fill_opacity=0.3,
                line_weight=0.25)



# continental_boundaries = (
# (25.0,-70.0),
# (50.0,-120.0)
# )

washington_y = 38.904978
washginton_x = -77.039658

washington_boundaries = (
(washington_y-0.15, washginton_x-0.15),
(washington_y+0.15, washginton_x+0.15)
)
    
# wider_map.fit_bounds(continental_boundaries)

wider_map.fit_bounds(washington_boundaries)
    
wider_map.save(outfile='wider_alpha.html')
```
