```python
from numpy import pi, cos, sin, sqrt
import matplotlib.pyplot as plt

"""function hex_to_pixel(hex):
    x = size * 3/2 * hex.q
    y = size * sqrt(3) * (hex.r + hex.q/2)
    return Point(x, y)"""



class hexagon:
    def __init__(self, coords, size=0.025):
        # there will be no coordinate reference system at the 
        # hex level, that needs to be handled by the grid        
        self.x_loc = coords[0]
        self.y_loc = coords[1]
        self.z_loc = -self.x_loc -self.y_loc
        self.size = size
        self.self_center()
        self.create_vertices()
    def self_center(self):
        self.x_center = self.size * 1.5 * self.x_loc
        self.y_center = self.size * sqrt(3) * (self.y_loc + self.x_loc/2)        
    def create_vertices(self):
        # I am only interested in making "flat top" layouts
        self.vertices = tuple(self.vertex(n) for n in range(6))
    def vertex(self, n):
        angle = n*pi/3
        i = self.size*cos(angle) + self.x_center
        j = self.size*sin(angle) + self.y_center
        return(i,j)
```

```python
import pandas as pd

class hexagonal_grid_axial:
    def __init__(self, width, size=0.025):
        self.size=size
        self.hex_frame = self.hex_dataframe_creator(width)
#         self.hexes = tuple(tuple(hexagon(axial_to_cube(x,y),size) for y in range(-x-3,width+1)) for x in range(-width,width+1))
    def set_crs(self, crs):
        pass
    def hex_dataframe_creator(self,w):
        array = [[]]
        for x in range(w+1):
            for y in range(2*w-x+1):
                array.append([(x,y-w),(x),(y-w),None, None])
        for x in range(-w,0):
            for y in range(2*w+x+1):
                array.append([(x,y+w),x,(y+w),None, None])
        result =  pd.DataFrame(array[1:], columns=['index','x','y', 
                                                   'hex_object', 
                                                  'interior points'])
        for row in result['hex_object'].index:
            result.loc[row, 'hex_object'] = hexagon((result.loc[row,'index']),self.size)
        return result.set_index('index')
        
base_frame = hexagonal_grid_axial(5,1)

base_frame.hex_frame.head()
        
```

```python
def hex_round(x,y):
    z = -x -y
    x_d, x = modf(x)
    y_d, y = modf(y)
    z_d, z = modf(z)
    
    if x_d>y_d and x_d>z_d:
        x = -y-z
    elif y_d>z_d:
        y = -x-z
    else:
        z=-x-y
        
    return x,y,z
```

```python
ssq = sqrt(3)/3

def point_loc_to_hex_loc(x_h, y_h, size):
    x = (x_h * 2/3)/ size
    y = ((-x_h/3) + (ssq * y_h)) / size
    return hex_round(x,y)
```

```python
def hex_center(x_row, y_row, size):
    x = size * 3/2 * x_row
    y = size * (3**0.5) * (y_row + x_row/2)
    return x, y
```
